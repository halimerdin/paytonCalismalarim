# Dosya okuma Yöntemleri

# For döngüsü ile okuma

```python
file=open("bilgiler.txt","r",encoding="utf-8")
for i in file:
    print(i)
file.close()
```

```python
file=open("bilgiler.txt","r",encoding="utf-8")
for i in file:
    print(i, end="")
file.close()
```

```python
file=open("bilgiler.txt","r",encoding="utf-8")
içerik=file.read()
print("dosyanın içeriği...")
print(içerik)
```

```python
file=open("bilgiler.txt","r",encoding="utf-8")
içerik=file.read()
print("dosyanın içeriği...")
print(içerik)
print("dosyanın içeriği 2:")
içerik2=file.read()
print(içerik2)
```

2.defa okuma işleminin yapılmamasının sebebi

ilk okumada file değişkenimiz (imleç görevi gödrdüğü için) dosyamızın sonuna gitti ve orda kaldı

```python
file=open("bilgiler.txt","r",encoding="utf-8")
print(file.readline())
file.close()
```

```python
file=open("bilgiler.txt","r",encoding="utf-8")
print(file.readline())
print(file.readline())
print(file.readline())
print(file.readline())
print(file.readline())
print(file.readline())
file.close()
```

readline() fonksiyonu okuyacak stringler kalmayınca sadce boş stringler döndürüyor


```python
file=open("bilgiler.txt","r",encoding="utf-8")
print(file.readline())
print(file.readline())
print(file.readline())
print(file.readline())
print("-------------")
print(file.readline())
print(file.readline())
file.close()
```
