0-7 aras toplamda 8 adet indeks var. istenilen indekse bir tane satr eklenirse istenilen sonuca ulalabilir.

yeni oluan listeyi butun halinde write fonksiyonu ile yazdrarak ilemi tamamlarz


```python
with open("bilgiler.txt","r+",encoding="utf-8") as file:
    liste=file.readlines()
    liste.insert(1,"tamamda neden ordasn\n")
    file.seek(0)
    for i in liste:
        file.write(i)
```

```python
with open("bilgiler.txt","r+",encoding="utf-8") as file:
    print(file.read())
```

for dgs ile btn listeyi yazmak yerine pythonda bulunan **writelines()** ile yazlabilir.

**writelines()** iine bir tane lsite alarak dosyamza listemizin her bir elemann yazar.



```python
with open("bilgiler.txt","r+",encoding="utf-8") as file:
    liste=file.readlines()
    liste.insert(1,"orta 2 fizik\n")
    file.seek(0)
    file.writelines(liste)
```

```python
with open("bilgiler.txt","r+",encoding="utf-8") as file:
    print(file.read())
```
