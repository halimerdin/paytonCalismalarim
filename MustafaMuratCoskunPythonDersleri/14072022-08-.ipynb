```python
class Yazılımcı():
    def __init__(a,isim,soyisim,numara,maas,diller):
        a.isim=isim
        a.soyisim=soyisim
        a.numara=numara
        a.maas=maas
        a.diller=diller
        
    def bilgilerigöster(a):
        print("""
        yazılımcı öğesinin özellikleri
        
        isim: {}
        
        soyisim: {}
        
        numara: {}
        
        maas: {}
        
        bildiği diller: {}
        
        """.format(a.isim,a.soyisim,a.numara,a.maas,a.diller))
    def zamYap(a,zamMiktarı):
        print("zam yapılıyor...")
        
        a.maas+=zamMiktarı
    def dilEkle(a,yeniDil):
        print("dil ekleniyr...")
        a.diller.append(yeniDil)
```

```python
class Çalışan():
    def __init__(self,isim,maaş,departman):
        print("çalışan sınıfının init fonksiyonu")
        """
        bu ifadenin ekrandagözükmesi init fonksiyonunun çalışması anlamına 
        geliyor        
        """        
        self.isim=isim
        self.maaş=maaş
        self.departman=departman
        
    def bilgileriGoster(self):
        print("çalışan sınıfının bilgileri .......")
        print("isim: {}\nmaaş:{} \ndepartman: {}".format(self.isim,self.maaş,self.departman))
    
    def departmanDegistir(self,yeniDepartmanIsmi):
        print("departman değiştiriliyor")
        self.departman=yeniDepartmanIsmi
```
