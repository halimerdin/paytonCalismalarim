0-7 arası toplamda 8 adet indeks var. istenilen indekse bir tane satır eklenirse istenilen sonuca ulaşılabilir.

yeni oluşan listeyi butun halinde write fonksiyonu ile yazdırarak işlemi tamamlarız


```python
with open("bilgiler.txt","r+",encoding="utf-8") as file:
    liste=file.readlines()
    liste.insert(1,"tamamda neden ordasın\n")
    file.seek(0)
    for i in liste:
        file.write(i)
```

```python
with open("bilgiler.txt","r+",encoding="utf-8") as file:
    print(file.read())
```

for dögüsü ile bütün listeyi yazmak yerine pythonda bulunan **writelines()** ile yazılabilir.

**writelines()** içine bir tane lsite alarak dosyamıza listemizin her bir elemanını yazar.



```python
with open("bilgiler.txt","r+",encoding="utf-8") as file:
    liste=file.readlines()
    liste.insert(1,"orta 2 fizik\n")
    file.seek(0)
    file.writelines(liste)
```
